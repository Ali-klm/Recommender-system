```python
import pandas as pd
import hazm
```

```python
df = pd.read_csv("../data/news_data.csv")
df.head()
```

```python
data = df.copy()
data.head()
```

```python
import re
from hazm import Normalizer
from hazm import WordTokenizer
from hazm import Lemmatizer
from hazm import stopwords_list


lemmatizer = Lemmatizer()
normalizer = Normalizer()
tokenizer = WordTokenizer()
stop_words = set(stopwords_list())


def persian_text_preprocessing(text):
    # نرمال‌سازی متن
    text = normalizer.normalize(text)

    # حذف علائم نگارشی و اعداد
    text = re.sub(r"[^\w\s]", " ", text)  # حذف علائم نگارشی
    text = re.sub(r"\d+", " ", text)  # حذف اعداد
    text = re.sub(r"\s+", " ", text)  # حذف فاصله‌های اضافی

    # توکن‌سازی
    tokens = tokenizer.tokenize(text)

    # حذف کلمات توقف
    tokens = [word for word in tokens if word not in stop_words]

    # ریشه‌یابی
    lemmas = [lemmatizer.lemmatize(word) for word in tokens]

    # بازترکیب توکن‌ها به متن نهایی
    return " ".join(lemmas)
```

```python
data["content"] = data["content"].apply(persian_text_preprocessing)
```

```python
data["content"].sample(5).values
```

```python
data["tag"] = data["label"] + " " + data["content"]
```

```python
data["tag"].sample(5).values
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# تعریف بردارساز TF-IDF با بیشینه ۵۰۰۰ ویژگی
tfidf = TfidfVectorizer(max_features=5000)

# اعمال بردارسازی روی ستون tag
tfidf_matrix = tfidf.fit_transform(data["tag"])
```

```python
print("شکل ماتریس TF-IDF:", tfidf_matrix.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# محاسبه شباهت کسینوسی بین بردارهای TF-IDF
similarity_matrix = cosine_similarity(tfidf_matrix)
```

```python
print("شکل ماتریس شباهت:", similarity_matrix.shape)

# مشاهده بخشی از ماتریس (مثلاً 5x5 اول)
import numpy as np

np.set_printoptions(precision=3, suppress=True)
print(similarity_matrix[:5, :5])
```

```python
def find_similar_news(news_index, similarity_matrix, data, top_n=5):
    if news_index < 0 or news_index >= len(data):
        print("شماره خبر معتبر نیست.")
        return []

    similarities = similarity_matrix[news_index]
    similar_indices = similarities.argsort()[::-1]
    similar_indices = [i for i in similar_indices if i != news_index]
    top_similar = similar_indices[:top_n]

    return top_similar
```

```python
print(df["content"][2])
```

```python
suggestion_result = find_similar_news(2, similarity_matrix, data)

print("\n✅ اخبار پیشنهادی مشابه با خبر شماره 2:\n")
for idx in suggestion_result:
    print(f"🔸 خبر مشابه #{idx}:\n{df.iloc[idx]['content'][:300]}...")
    print("—" * 80)
```

```python
# پیدا کردن ۵ خبر مشابه با خبر شماره ۴۶
similar_indices = find_similar_news(46, similarity_matrix, data, top_n=5)
```

```python
# ساخت دیتافریم خروجی شامل label و index
submission = pd.DataFrame(
    {"label": df.loc[similar_indices, "label"].values, "index": similar_indices}
)
```
