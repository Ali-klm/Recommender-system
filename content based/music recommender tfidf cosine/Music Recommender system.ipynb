```python
import numpy as np
import pandas as pd
import sklearn
```

```python
data = pd.read_csv("../data/music_data.csv")
data.head()
```

```python
# انتخاب ستون‌های مهم برای سیستم پیشنهاددهنده
selected_features = [
    "track_name",
    "artists",
    "track_genre",
    "danceability",
    "energy",
    "valence",
    "tempo",
    "acousticness",
    "instrumentalness",
]

# ساخت یک نسخه‌ی جدید از داده با فقط ویژگی‌های مورد نظر
df = data[selected_features].copy()
df.head()
```

```python
df.info()
```

```python
# بررسی مقدار داده‌های گمشده در هر ستون
df.isnull().sum()
```

```python
import re


def preprocessing_text(text):
    # تبدیل به حروف کوچک
    text = text.lower()

    # حذف کاراکترهای غیر حرف یا عدد (مثل !@#$...)
    text = re.sub(r"[^a-z0-9\s]", "", text)

    # حذف فاصله‌های اضافی
    text = re.sub(r"\s+", " ", text).strip()

    return text
```

```python
df["track_name_clean"] = df["track_name"].apply(preprocessing_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# پیش‌پردازش روی هر ستون متنی
df["track_name_clean"] = df["track_name"].apply(preprocessing_text)
df["artists_clean"] = df["artists"].apply(preprocessing_text)
df["genre_clean"] = df["track_genre"].apply(preprocessing_text)

# ترکیب متون در یک ویژگی جدید
df["text_feature"] = (
    df["track_name_clean"] + " " + df["artists_clean"] + " " + df["genre_clean"]
)

# بردارسازی با TF-IDF
vectorizer = TfidfVectorizer()
X_text = vectorizer.fit_transform(df["text_feature"])
```

```python
# TO DO
from sklearn.preprocessing import MinMaxScaler

numeric_cols = [
    "danceability",
    "energy",
    "valence",
    "tempo",
    "acousticness",
    "instrumentalness",
]
scaler = MinMaxScaler()
X_numeric = scaler.fit_transform(df[numeric_cols])
```

```python
from scipy.sparse import hstack
from scipy.sparse import csr_matrix  # تبدیل X_numeric به sparse

# تبدیل ویژگی‌های عددی به sparse matrix
X_numeric_sparse = csr_matrix(X_numeric)

# ترکیب ویژگی‌های متنی و عددی
X_combined = hstack([X_text, X_numeric_sparse])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# محاسبه ماتریس شباهت کسینوسی بین همه آهنگ‌ها
cosine_sim = cosine_similarity(X_combined)
```

```python
def recommender(music_name):
    # پیدا کردن ایندکس آهنگ بر اساس نام (پیش‌پردازش نام برای اطمینان)
    music_name = preprocessing_text(music_name)

    # پیدا کردن ردیف آهنگ مورد نظر در دیتافریم
    idx = df[df["track_name_clean"] == music_name].index

    if len(idx) == 0:
        return f"آهنگ '{music_name}' در دیتاست یافت نشد."

    idx = idx[0]  # گرفتن اولین ایندکس در صورت چندگانه بودن

    # گرفتن نمرات شباهت برای این آهنگ
    sim_scores = list(enumerate(cosine_sim[idx]))

    # مرتب‌سازی بر اساس امتیاز شباهت، به جز خودش (index == idx)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # حذف خودش از لیست (نمره شباهت 1.0 برای خودش)
    sim_scores = [score for score in sim_scores if score[0] != idx]

    # گرفتن 21 آهنگ مشابه اول
    top_21 = sim_scores[:21]

    # استخراج نام آهنگ‌ها بر اساس ایندکس‌ها
    recommended_tracks = [df.iloc[i[0]]["track_name"] for i in top_21]

    return recommended_tracks
```

```python
selection = ["Ilahi", "Loony", "This Is Your Life", "Solamente una Vez", "In Da Getto"]
```

```python
selection = ["Ilahi", "Loony", "This Is Your Life", "Solamente una Vez", "In Da Getto"]

# دیکشنری برای ذخیره نتایج
recommendations_dict = {}

for song in selection:
    recommendations = recommender(song)
    # اگر خروجی پیام خطا بود، اون رو با لیست خالی جایگزین کن
    if isinstance(recommendations, str):
        recommendations = [None] * 21
    recommendations_dict[song] = recommendations

# ساخت دیتافریم از دیکشنری
submission = pd.DataFrame(recommendations_dict)

submission
```
